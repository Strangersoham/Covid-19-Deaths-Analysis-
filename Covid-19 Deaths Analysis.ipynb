### Introduction 

COVID-19 is an infectious disease caused by the Corona Virus, biologically known as severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The disease was first identified in Wuhan, the capital of China's Hubei province in December 2019 and has spread all over the world since then. As of writing this, on 26th April 2020, 21:57 IST, there are 2.92 million confirmed cases throughout the world and has resulted in 204,000 deaths according to Google.

In this notebook, I will take a look at the current situation in India. We will take a look at the regions which are most  effected by the outbreak and how numbers have steadily climbed in the country .



```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
covid = pd.read_csv('covid_19_india.csv')
```

```python
MH = covid.loc[covid['State/UnionTerritory'] == "Maharashtra"]
TS = covid.loc[covid['State/UnionTerritory'] == "Telengana"]
UP = covid.loc[covid['State/UnionTerritory'] == "Uttar Pradesh"]
```

```python
plt.plot(TS.Date ,TS.Deaths,color= "red")
plt.xticks(TS.Date[::10])
plt.title('Increase in deaths due to Covid 19')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.show()
```

```python
plt.plot(MH.Date ,MH.Deaths,color= "blue")
plt.xticks(MH.Date[::10])
plt.title('Increase in deaths due to Covid 19')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.show()
```

```python
plt.plot(UP.Date ,UP.Deaths,color= "green")
plt.xticks(UP.Date[::10])
plt.title('Increase in deaths due to Covid 19')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.show()
```

```python
plt.plot(TS.Date ,TS.Deaths,color= "red", label="TS")
plt.plot(UP.Date ,UP.Deaths ,color= "green",label="UP")
plt.plot(MH.Date ,MH.Deaths ,color= "blue",label="MH")
plt.xticks(TS.Date[::10])
plt.title('Increase in deaths due to Covid 19')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.legend()
plt.show()
```

```python
covid.hist(column='Deaths')
plt.show()
```

```python
plt.figure(figsize=(20,10))
covid['State/UnionTerritory'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
from matplotlib import rcParams
rcParams['figure.figsize']= 10,6
plt.grid(True)
plt.plot(TS.Date ,TS.Confirmed,color= "red", label="TS")
plt.plot(UP.Date ,UP.Deaths ,color= "green",label="UP")
plt.plot(MH.Date ,MH.Deaths ,color= "blue",label="MH")
plt.xticks(TS.Date[::10])
plt.title('Increase in deaths due to Covid 19')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.legend() 
plt.show()
```
